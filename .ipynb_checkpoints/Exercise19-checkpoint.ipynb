# Decision Tree

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn.model_selection import train_test_split
```



## Read "bank-full.csv"

```python
data = pd.read_csv("./bank-full.csv", sep=';')
df = pd.DataFrame(data)
df
```

## Check the distribution of labels (‘yes’, ‘no’) are distributed.


```python
sns.countplot(x=df["y"])
```

## Counts of "yes" and "no" with "age"

```python
df.groupby('age')['y'].value_counts()
```

## Correlation between the data

```python
df.corr(numeric_only= True)
```

Plot the heatmap

```python
sns.heatmap(df.corr(numeric_only = True))
```

## Convert categorical data into numerical

```python
reset = {"yes": 1, "no": 0, "unknown": None}
df = df.replace({"housing": reset})
df = df.replace({"loan": reset})
df = df.replace({"education": {"primary": 1, "secondary": 2, "tertiary": 3, "unknown": None}})
df = df.replace({"marital": {"single": 1, "married": 2, "divorced": 3, "unknown": None}})
df = df.replace({"contact": {"cellular": 1, "telephone": 0, "unknown": None}})
df = df.replace({"y": {"yes": 1, "no": 0, "unknown": None}})
df = df.dropna()
df
```

Next step is to select features and labels

```python

feature_cols = ['marital', 'education', 'balance', 'duration', 'housing', 'loan', 'contact', 'campaign', 'age']
X = df[feature_cols] # Features
y = df.y # Target variable
```

Drop "poutcome"

```python
del df['poutcome']
df
```

## Split the data into train and test

```python
import pandas as pd


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) 
```

# Applying Decision Tree Classifier:

Next, I created a pipeline of StandardScaler (standardize the features) and DT Classifier (see a note below regarding Standardization of features). We can import DT classifier as from sklearn.tree import DecisionTreeClassifier from Scikit-Learn. To determine the best parameters (criterion of split and maximum tree depth) for DT classifier, I also used Grid Search Cross Validation. The code snippet below is self-explanatory.

```python
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics, tree

# Create Decision Tree classifer object
clf = DecisionTreeClassifier(max_depth = 3)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

## To display

```python
plt.figure(figsize=(10, 6))
tree.plot_tree(clf, filled= True)
plt.show()
```

The number of nodes and the maximum depth

```python
print(clf.tree_.node_count, clf.tree_.max_depth)
```

## Accuracy measurement

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

# Prediction

```python
y_pred
```

# Grid Search

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

pipe = Pipeline(steps=[('dec_tree', clf)])
criterion = ['gini', 'entropy']
max_depth = [2,4,6,8,10,12]

parameters = dict(dec_tree__criterion = criterion, dec_tree__max_depth = max_depth)

grid_search = GridSearchCV(pipe, parameters)

grid_search.fit(X_train, y_train)
```

## Display the best features

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=100)
classifier.fit(X_train, y_train)

feature_importances_df = pd.DataFrame(
    {"feature": list(X.columns), "importance": classifier.feature_importances_}
).sort_values("importance", ascending=False)

# Display
feature_importances_df
```

## Run DecisionTreeClassifier using the obtained features

```python
optimized = DecisionTreeClassifier(criterion='gini', max_depth=2)
optimized.fit(X_train, y_train)
```

## Concat train test results

```python
y_pred_train = optimized.predict(X_train)
y_pred_test = optimized.predict(X_test)
y_train_np = y_train.to_numpy().reshape(len(y_train), 1)
y_test_np = y_test.to_numpy().reshape(len(y_test), 1)

result_train = np.concatenate((y_pred_train.reshape(-1, 1), y_train_np), axis=1)
result_test = np.concatenate((y_pred_test.reshape(-1, 1), y_test_np), axis=1)

print("Training:")
print(result_train)

print("Test:")
print(result_test)
```
