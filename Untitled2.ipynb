```python
import pandas as pd

pd.set_option('display.max_columns', None)

null_counts = df.isnull().sum()

columns_to_drop = null_counts[null_counts > 100].index
df_cleaned = df.drop(columns = columns_to_drop)
unique_counts = df_cleaned.nunique()
nunique_columns_filter = (unique_counts == 0) | (unique_counts == 1) | (unique_counts == 999)
columns_to_keep = df_cleaned.columns[nunique_columns_filter]
df_cleaned = df_cleaned.drop(columns=columns_to_keep)

len(df_cleaned.columns)
df_cleaned
```

```python
df_cleaned.info()
```

```python
object_columns = df_cleaned.select_dtypes(include='object').columns

# Check for null values in object-type columns (empty strings or NaN)
columns_with_null = df_cleaned[object_columns].isnull().any()
columns_with_null
```

```python
columns_with_null = ['Series', 'BadgeDescription', 'VIN', 'EngineNum', 'Colour', 'Compliance_Date', 'EngineConfigurationDescription']

# Remove rows with null values in the specified columns
df_cleaner = df_cleaned.dropna(subset=columns_with_null, how='any')
df_cleaner
```

```python
df_cleaner.describe(include='object')
```

```python
null_counts = df_cleaner.isnull().sum()

for i in null_counts[null_counts > 0].index:
    if df_cleaner[i].dtype != 'O':
        df_cleaner[i].fillna(df_cleaner[i].median(), inplace=True)
df_cleaner.info()
```

```python
df_cleaner.corr()
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

for i in df_cleaner.columns:
    if df_cleaner[i].dtype == 'O':
        df_cleaner[i] = le.fit_transform(df_cleaner[i])
df_cleaner
```
